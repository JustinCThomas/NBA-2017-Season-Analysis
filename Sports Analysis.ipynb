```python
import csv 

# Columns to use: ["Year", "Player", "Pos", "Age", "Tm" "G", "AST", "STL", "BLK", "TOV", "PTS"]
mappings = {"Year": 1, "Player": 2, "Pos": 3, "Age": 4, "Tm": 5, "G": 6, 
            "AST": 47, "STL": 48, "BLK": 49, "TOV": 50, "PTS": 52}

def season_data_np_arr(season=2017):
    data = []
    with open("nba-players-stats/Seasons_Stats.csv", "r") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=",")
        count = 0
        for row in csv_reader:

            if row[mappings["Player"]].upper() in player_names and row[mappings["Year"]] == str(season):
                data += [[row[mappings["Year"]], row[mappings["Player"]], row[mappings["Pos"]],
                      row[mappings["Age"]], row[mappings["Tm"]], row[mappings["G"]],
                      row[mappings["AST"]], row[mappings["STL"]], row[mappings["BLK"]],
                      row[mappings["TOV"]], row[mappings["PTS"]]]]
            
    return np.array(data)
    
season2014 = season_data_np_arr(2014)
season2015 = season_data_np_arr(2015)
season2016 = season_data_np_arr(2016)
season2017 = season_data_np_arr(2017)
           
```

```python
def season_data_pd_df(season=2017):
    rows_list = []
    with open("nba-players-stats/Seasons_Stats.csv", "r") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=",")
        count = 0
        for row in csv_reader:
            data = {}
            
            if row[mappings["Player"]].upper() in player_names and row[mappings["Year"]] == str(season):
                row_data = {}
                row_data["Year"] = int(row[mappings["Year"]])
                row_data["Player"] = row[mappings["Player"]]
#                 row_data["Player"] = row[mappings["Player"]].upper()
                row_data["Pos"] = row[mappings["Pos"]]
                row_data["Age"] = int(row[mappings["Age"]])
                row_data["Tm"] = row[mappings["Tm"]]
                row_data["G"] = int(row[mappings["G"]])
                row_data["AST"] = int(row[mappings["AST"]])
                row_data["STL"] = int(row[mappings["STL"]])
                row_data["BLK"] = int(row[mappings["BLK"]])
                row_data["TOV"] = int(row[mappings["TOV"]])
                row_data["PTS"] = int(row[mappings["PTS"]])
                 

                data.update(row_data)

                rows_list.append(data)

    return pd.DataFrame(rows_list)

            
df2017 = season_data_pd_df(2017)
```

```python
def player_data_pd_df(player_name):
    rows_list = []
    with open("nba-players-stats/Seasons_Stats.csv", "r") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=",")
        count = 0
        for row in csv_reader:
            data = {}
            
            if row[mappings["Player"]].upper() == player_name.upper():
                row_data = {}
                row_data["Year"] = int(row[mappings["Year"]])
                row_data["Player"] = row[mappings["Player"]]
#                 row_data["Player"] = row[mappings["Player"]].upper()
                row_data["Pos"] = row[mappings["Pos"]]
                row_data["Age"] = int(row[mappings["Age"]])
                row_data["Tm"] = row[mappings["Tm"]]
                row_data["G"] = int(row[mappings["G"]])
                row_data["AST"] = int(row[mappings["AST"]])
                row_data["STL"] = int(row[mappings["STL"]])
                row_data["BLK"] = int(row[mappings["BLK"]])
                row_data["TOV"] = int(row[mappings["TOV"]])
                row_data["PTS"] = int(row[mappings["PTS"]])
                 

                data.update(row_data)

                rows_list.append(data)

    return pd.DataFrame(rows_list)

axis = plt.gca()

KD = player_data_pd_df("Kevin Durant")
plt.title("Kevin Durant Stats Progression")
KD.plot(kind="line", x="Year", y="STL", ax=axis)
KD.plot(kind="line", x="Year", y="TOV", ax=axis)
KD.plot(kind="line", x="Year", y="BLK", ax=axis)
KD.plot(kind="line", x="Year", y="AST", ax=axis)
plt.show()

KD.plot(kind="line", x="Year", y="PTS", c="#0055FF")
plt.title("Kevin Durant Stats Progression")
plt.show()


KD.plot(kind="line", x="Year", y="G", c="c", label="Games")
plt.title("Kevin Durant Games per Season")
plt.show()
```
